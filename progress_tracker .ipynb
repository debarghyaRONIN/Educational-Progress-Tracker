 The first code is reading the csv file using pandas library and reading the whole data from the csv file of YEAR1, YEAR2, YEAR3, YEAR4 and by using the matplotlib.pyplot abbreviated as plt, we have plotted a bargraph using df_selected(kind='bar') and df_selected is a variable which is used to take the data from df variable of Name,Y1,Y2. Y1,Y2,Y3,Y4 is used to show the marks each person got in a year.

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_student_progress(csv_file_path):
    df = pd.read_csv(csv_file_path)
    df_selected = df[['Name','Y1','Y2']]
    #df2 = df.head()
    df_selected.set_index('Name',inplace=True)
    df_selected.plot(kind='bar')
    plt.rcParams["figure.figsize"] = (20,10)
    plt.show()

plot_student_progress("/content/dataset.csv")
```

This code makes a more pleasing line graph using seaborn which is made over matplotlib,it is amazing for plotting graphs with less lines of code. the .rcparams() is used here to define the zoom size of the graph.

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_student_progress(csv_file_path):
    df = pd.read_csv(csv_file_path)
    df2= df.head()

    df2.set_index('Name', inplace=True)

    sns.lineplot(data=df2.T, linewidth=4)  # Transpose the DataFrame for seaborn lineplot
    plt.rcParams["figure.figsize"] = (20, 10)
    plt.show()

plot_student_progress("/content/dataset.csv")
```

```python
#code for selective data used for getting the dataplot for year 1 and year 2 from the dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_student_progress(csv_file_path):
    df = pd.read_csv(csv_file_path)


    df.set_index('Name', inplace=True)

    # Selecting columns 'Y1' and 'Y2' for all names
    df_selected = df[['Y1', 'Y2']]

    #Lineplot using Seaborn
    sns.lineplot(data=df_selected, linewidth=5)

    # Adjusting plot size using rcParams by putting (x,y) parameters.
    plt.rcParams["figure.figsize"] = (2000, 10)


    plt.show()

#function execution on datasets
plot_student_progress("/content/dataset.csv")
```
